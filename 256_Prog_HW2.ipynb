```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import cosine
```

```python
##read from file##
train_data = pd.DataFrame(pd.read_csv('train.csv'))
test_data = pd.DataFrame(pd.read_csv('test.csv'))
```

```python
##preserve 'ID' column for output
output = test_data['ID']
```

```python
##drop useless columns##
train_data = train_data.drop(['Timestamp','ID','Age?','Gender','Preferred Genres'],1)
test_data =test_data.drop(['ID','Age?','Gender','Preferred Genres'],1)
```

```python
##drop whole row if "Despicable Me" is note rated 
##-> because that cannot be used to calculate the collaborated filter
train_data = train_data[train_data['Rating [Despicable Me]']>0]
```

```python
##normalized by mean in each row
mean = train_data.mean(1)
train_data = train_data.sub(mean,axis=0)
mean_test = test_data.mean(1)
test_data = test_data.sub(mean_test,axis=0)
```

```python
##fill in NaN with 0. 0 can help simularity calculation and no harm
train_data = train_data.fillna(0)
test_data = test_data.fillna(0)
##reset index of train_data, let the FOR Loop calculation easier
train_data = train_data.reset_index(drop=True)
```

```python
##calculate the simularity and put it into "sim_matrix" array
##i: iteration in test_data
##j: iteration in train_data
sim_matrix = np.zeros((len(test_data.index), len(train_data.index)))
for i in range(0,len(test_data.index)):
    for j in range(0,len(train_data.index)):
        sim_matrix[i][j] = 1-cosine(train_data.loc[j,:],test_data.loc[i,:])
        
sim_matrix
```

```python
##using user-base collaberative filter scheme to calculate the ratings
##for each test item(i), summating positive simularity items(j) from trainning data(j=0,1,2,...,total valid data)
for i in range(0,len(test_data.index)):
    rating = np.zeros(1)
    sumsim = np.zeros(1)
    for j in range(0,len(train_data.index)):
        ##choose only positive sumularity items
        if(sim_matrix[i][j]>0):
            ##summating "ratings *simularity" in traiining data 
            rating += sim_matrix[i][j]*train_data.loc[j,'Rating [Despicable Me]']
            ##also summating simularities
            sumsim += sim_matrix[i][j]
    ##for each test item(i), calculate "(ratings*simularity)/simularity_sum" as the predict rating
    test_data.loc[i,'Rating [Despicable Me]']= rating / sumsim
```

```python
##fill in NaN with 0. NaN means there is no enough simularity info to calculate its rating
test_data = test_data.fillna(0)
```

```python
##add mean value back, becoming the normal rating scores
test_data = test_data.add(mean_test,axis=0)
```

```python
##concatenate ID and rating as example format
output = pd.concat([output, test_data['Rating [Despicable Me]']],axis=1)
```

```python
output.to_csv('rating.csv', index=False)
```
